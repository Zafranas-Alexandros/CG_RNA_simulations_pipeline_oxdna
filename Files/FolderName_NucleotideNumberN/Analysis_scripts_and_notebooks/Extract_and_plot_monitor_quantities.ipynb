Before running any of the cells below, one has to create the output_file.xvg by means of "script_to_clean_logfile_lammps.sh". After running the script, one has to also manually clean the output file, specifically: delete the first rows which don't contain relevant epot information and make sure that the number of digits on the first column -timeframe info- is correct.

```python
import numpy as np
#Use numpy to load the data from files
t1,etot = np.loadtxt("etot.xvg", unpack=True)
t2,epot = np.loadtxt("epot.xvg", unpack=True)
t3,temp = np.loadtxt("temp.xvg", unpack=True)

```

```python
import matplotlib.pyplot as plt
#plot quantities of interest over simulation time
fig = plt.figure(figsize=(10,10))
gs = fig.add_gridspec(3, hspace=0.1)
axs = gs.subplots(sharex=True)
axs[0].plot(t1[:-1], etot[:-1], 'g-')
axs[0].set_ylabel("Total Energy")
axs[1].plot(t2[:-1], epot[:-1], 'b-')
axs[1].set_ylabel("Potential Energy")
axs[2].plot(t3, temp*3000, 'r-')
axs[2].set_ylabel("Temperature")
axs[2].set_xlabel('Time')
avg_T = np.mean(temp)
axs[2].hlines(avg_T*3000, xmin=min(t3), xmax=max(t3), colors='black', linestyle = "--")
print(avg_T*3000)
```
