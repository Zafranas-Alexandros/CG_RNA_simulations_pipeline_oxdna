```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

#Use numpy to load the data from a file 
d,g = np.loadtxt("rgyr.dat", unpack=True)

# pretty standard plot code
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.plot(d, g, 'g-')
ax.set_xlabel('Frame')
ax.set_ylabel(r'radius of gyration')
```

```python
#Use numpy to load the data from a file 
d,g = np.loadtxt("rmsd.dat", unpack=True)

# pretty standard plot code
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.plot(d, g, 'm-')
ax.set_xlabel('Frame')
ax.set_ylabel(r'RMSD')
```

```python
#Use numpy to load the data from a file 
d = np.loadtxt("rmsf.dat", unpack=True)

g = list(range(len(d)))

# pretty standard plot code
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.plot(g, d, 'r-')
ax.set_xlabel('Frame')
ax.set_ylabel(r'RMSF')
```

```python
#Use numpy to load the data from a file 
d,g = np.loadtxt("MSD.dat", unpack=True)

G = np.array(g)
GG = np.log10(G)

D = np.array(d)
DD = np.log10(D)

# pretty standard plot code
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111)
ax.plot(DD, GG, 'b-')
ax.set_xlabel('Frame')
ax.set_ylabel(r'MSD')
```
